###Mean Squared Error (MSE )

```python
def mse_cal(y,h):
  return np.mean((y-h)**2)
```

# ทำความรู้จัก Loss function

ขั้นตอนต่อไป เราต้องการจะรู้ว่า ณ ค่า paremeter w ค่าต่าง ๆ นั้น ทำให้โมเดลของเรามีค่า error เท่าไร

เพื่อว่า ต่อไปเราจะทดลองปรับค่า parameter w แล้วจะได้ดูว่า การปรับค่านั้นช่วยให้โมเดลของเรามี error เปลี่ยนแปลงไปอย่างไร

จึงสร้างเป็น Loss function ขึ้นดัง code ข้างล่างนี้

```python
def Loss_func(w):
  h = w*x
  return mse_cal(y,h)
```

วิธีการการหาค่า optimal parameter ที่ simple ที่สุด ก็คือลองสุ่มค่า w ขึ้นมาจำนวนมาก แล้วทำลูปให้ทดลองทุกค่า w ที่สุ่มมา แล้วดูว่าค่า w ค่าไหน ที่ทำให้โมเดลมีค่า error ต่ำที่สุด

```python
#create lots of w for testing which w is the best
wList = np.arange(-200,240)/10
wList
```

```python
# simple optimal parameter finding method

# create empty list to record error for each W
mseList = []

# random wList
wList = np.arange(-200,240)/10

# loop for every w in wList
for w in wList:
  mse = Loss_func(w)
  mseList.append(mse)

# plot Loss function
py.scatter(wList, mseList)
```

```python
edat = pd.DataFrame()
edat['mse'] = mseList
edat['weight'] = wList

# find row that has the lowest mse
edat.sort_values(by='mse').head()
```

วิธีการนี้ก็คงใช้ได้กับเฉพาะกรณีที่มี parameter จำนวนน้อย ๆ แต่ในโลกความเป็นจริง โมเดลจะมีความซับซ้อนกว่านี้มาก จำนวน parameter ก็มีจำนวนมาก การจะมาสุ่มหา ก็คงต้องใช้เวลานานมาก หรือแทบเป็นไปไม่ได้เลยที่จะเจอค่า optimal parameter

วิธีการที่รวดเร็ว และเป็นวิธีการที่โมเดลส่วนใหญ่ในปัจจุบันใช้ในการหา optimal paremter เรียกว่า `Gradient descent algorithm`

# วิธีการหาค่า parameter ที่ดีที่สุดโดยใช้ Gradient descent

CostFunctionFormula.PNG

```python
# dj by dw
np.mean((h - y) * x)
```

```python
# initialize weight
w = -10000

mse = Loss_func(w)

# learning rate
alpha = 0.0001

# number of iteration
iter = 20

wList = [w]
mseList = [mse]
mse
```

##ทดลองปรับค่า w โดยใช้ gradient descent

```python
# initial w value = -10000
w
```

```python
for i in range(iter):
  h = w * x
  dj_by_dw = np.mean((h - y) * x)
  w = w - alpha * dj_by_dw
```

```python
w
```

##Plot กราฟดูการขยับเส้น h ตาม w ที่เปลี่ยนแปลงไป

```python
# re initialize w = -1000
w = -1000
```

```python
for i in range(iter):
  h = w * x
  dj_by_dw = np.mean((h - y) * x)
  w = w - alpha * dj_by_dw

  fig, (ax1, ax2) = py.subplots(1, 2)
  mseList.append(mse)
  wList.append(w)
  ax1.scatter(x, y)
  ax1.plot(x,x*w,c='r')
  ax2.scatter(wList, mseList)
  py.show()
```

##สร้าง function train_model เพื่อทดลองค่า hyperparameter ค่าต่าง ๆ

```python
def train_model(iter=20, w=-10000, alpha=0.0001):
  mse = Loss_func(w)
  wList = [w]
  mseList = [mse]
  for i in range(iter):
    h = w * x
    dj_by_dw = np.mean((h - y) * x)
    w = w - alpha * dj_by_dw
    fig, (ax1, ax2) = py.subplots(1, 2)
    mseList.append(mse)
    wList.append(w)
    ax1.scatter(x, y)
    ax1.plot(x,x*w,c='r')
    ax2.scatter(wList, mseList)
    py.show()
  return wList, mseList
```

```python
# ลอง train ให้นานขึ้นจาก 20 -> 50
w_list, loss_list = train_model(iter=50)
```
